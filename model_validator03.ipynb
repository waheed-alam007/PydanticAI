```python
from pydantic import BaseModel, EmailStr, model_validator
from typing import List, Dict

class Patient(BaseModel):

    name: str
    email: EmailStr
    age: int
    weight: float
    married: bool
    allergies: List[str]
    contact_details: Dict[str, str]

    # here we work not any single field but now i have  to work on the whole pydantic model no need to specify the field name 
    @model_validator(mode='after')
    def validate_emergency_contact(self):  
        # nnow make the method name validate_emergency_contact  now  in this method get two values clas and model due to this you have ton able to access any field 
        # in this if patint greater then 50 then must have the emergency conct number otherwise not get the patient if emergency contact has then return as it model
        if self.age > 50 and 'emergency' not in self.contact_details:
            raise ValueError('Patients older than 50 must have an emergency contact')
        return self

def update_patient_data(patient: Patient):

    print(patient.name)
    print(patient.age)
    print(patient.allergies)
    print(patient.married)
    print('updated')

patient_info = {
    'name': 'ali',
    'age': 65,
    'weight': 45.3,
    'married': True,
    'email': 'alkahr@gmail.com',
    'allergies': ['Pollen', 'Dust'],
    'contact_details': {'Email': 'alkahr@gmail.com', 'phone': '1234564','emergency':'4325617'}  
    # to build this raw data using the below pydantic object and then  and then pass this pydantic object to our function
}

patient1 = Patient(**patient_info)


update_patient_data(patient1)

# if you set in the dictionary patient infor age value=grter then the 50 then it will raise the error for this purpose we pass in the dictionary emergency nuber then it will genrate proper .
```
