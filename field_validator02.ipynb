**first use cse field-validator**

```python
from pydantic import BaseModel,EmailStr,AnyUrl,Field,field_validator
from typing import List,Dict,Optional,Annotated
class Patient(BaseModel):
    name:str
    email:EmailStr
    linkedin_url:AnyUrl                         
    age:int 
    weight:float
    married:bool 
    allergies:list[str] 
    contact_detail:Dict[str,str]

    # for this in emalst validator we have to create a method and make sure this method use with a decorator and iits name @field_validator ,here you tell this decorator apply in which field as in this case i want to apply in email,and you also tell the method you hhave to creat this method is class method @classmethod
    @field_validator('email')
    @classmethod
    # now we have to define our method and in this pass two this class and value in the value has ha email that will provided by patient and you check in the values has MCB/HBL.com avalible or not.
    def email_validator(cls,value) :
        # now we check the emali in value for this we have to make the a list in valid_dom and pas MCB,HBL
        valid_domain=['mcb.com','hbl.com']
        # we have to extract after the @ part as user provide email adef@gmail.com
        domain_name = value.split('@')[-1]    #-1 mean we need last part as gmail.com,mcb.com etc
        if domain_name not in valid_domain:
            raise ValueError('Not a valid domin')
        # otherwise i will return what ever  inside value
        return value



def insert_patient_data(patient:Patient):  
    print(patient.name)  
    print(patient.age)
    print(patient.weight)
    print(patient.married)
    print(patient.allergies)
    print(patient.email)
    print(patient.linkedin_url)
    print('inserted')

def update_patient_data(patient:Patient):     
    print(patient.name)
    print(patient.age)
    print(patient.weight)
    print(patient.married)
    print(patient.allergies)
    print(patient.email)
    print(patient.linkedin_url)
    print('updated')

patient_info={'name':'ali','age':22,'weight':45.3,'married':True,'linkedin_url':'http://linkedin.com/123', 'email':'alkahr@mcb.com', 'allergies':['Pollen','Dust'],'contact_detail':{'phone':'1234564'}}  #to build this raw data using the below pydantic object and then  and then pass this pydantic object to our function
patient1=Patient(**patient_info)

insert_patient_data(patient1)   
update_patient_data(patient1)


# always remeber field_validator is a class method @field_validator 
```
